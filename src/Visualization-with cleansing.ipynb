```python
#Libraries used for visualization
import numpy as np
import pandas as pd
import seaborn as sns
import re
import matplotlib.pyplot as plt
```

```python
# project filename
filename="CleanedDataset.csv"
# Load wiki plots data set
data = pd.read_csv(filename)
#Assigs different features to python variables
Directors=data["Director"]
Genres=data["Selected_Genre"]
Cast=data["Cast"]
Ethnicity=data["Origin/Ethnicity"]
Title=data["Title"]
Release_Year=data["Release Year"]
```

```python
#shows the Top 20 movie directors for this dataset
booleans=[]
for result in Directors:
    if not re.search('Unknown',result):
        booleans.append(True)
    else:
        booleans.append(False)
Filtered=pd.Series(booleans)
New_data=data[Filtered]
New_Directors=New_data["Director"]
sns.countplot(New_Directors, order=pd.value_counts(New_Directors)[:20].index, palette=sns.color_palette("seismic", 20))
plt.xlabel('Directors',fontsize=15)
plt.ylabel('No of movies directed',fontsize=15)
plt.title('Top Directors')
plt.xticks(rotation=90)
```

```python
#shows the Top 10 movies genres for this dataset
boolean=[]
for new_result in Genres:
    if not re.search('Unknown',new_result):
        booleans.append(True)
    else:
        booleans.append(False)
Filtered=pd.Series(booleans)
New_data=data[Filtered]
New_Genres=New_data["Selected_Genre"]
sns.countplot(New_Genres, order=pd.value_counts(New_Genres)[:10].index, palette=sns.color_palette("dark", 20))
plt.xlabel('Movie Genres',fontsize=15)
plt.ylabel('No of movie genres',fontsize=15)
plt.title('Most popular movie Genres')
plt.xticks(rotation=90)
```

```python
#shows the Top 10 movie origination for this dataset
sns.countplot(Ethnicity,order=pd.value_counts(Ethnicity)[:10].index, palette=sns.color_palette("dark", 20))
plt.xlabel('Movie Origin',fontsize=15)
plt.ylabel('No of movies of specific origin',fontsize=15)
plt.title('Origin Of Movies')
plt.xticks(rotation=90)
```

```python
#shows a time seies for the movies released from start to finish
Released_year_counts=data['Release Year'].value_counts().sort_index(ascending=True)
plt.plot(Released_year_counts)
plt.xlabel('Release Year',fontsize=15)
plt.ylabel('No of movies realeased',fontsize=15)
plt.title('Time series of movies ')
```

```python

Drama_Genre=(New_data[New_data['Genre'].str.contains("drama")])
searchfor = ['American','British','Bollywood','Tamil']
New_Ethnicity_Drama=Drama_Genre[Drama_Genre['Origin/Ethnicity'].str.contains('|'.join(searchfor))]
New_Ethnicity_Drama=(New_Ethnicity_Drama['Origin/Ethnicity'])
New_Directors_Drama=Drama_Genre['Director']
###
Comedy_Genre=(New_data[New_data['Genre'].str.contains("comedy")])
searchfor = ['American','British','Bollywood','Tamil']
New_Ethnicity_Comedy=Comedy_Genre[Comedy_Genre['Origin/Ethnicity'].str.contains('|'.join(searchfor))]
New_Ethnicity_Comedy=(New_Ethnicity_Comedy['Origin/Ethnicity'])
New_Directors_Comedy=Comedy_Genre['Director']
##
Horror_Genre=(New_data[New_data['Genre'].str.contains("horror")])
searchfor = ['American','British','Bollywood','Tamil']
New_Ethnicity_Horror=Horror_Genre[Horror_Genre['Origin/Ethnicity'].str.contains('|'.join(searchfor))]
New_Ethnicity_Horror=(New_Ethnicity_Horror['Origin/Ethnicity'])
New_Directors_Horror=Horror_Genre['Director']
##
Action_Genre=(New_data[New_data['Genre'].str.contains("action")])
searchfor = ['American','British','Bollywood','Tamil']
New_Ethnicity_Action=Action_Genre[Action_Genre['Origin/Ethnicity'].str.contains('|'.join(searchfor))]
New_Ethnicity_Action=(New_Ethnicity_Action['Origin/Ethnicity'])
New_Directors_Action=Action_Genre['Director']
```

```python
#creates a 2 by 2 subplot for the origin/ethinicty based on the different genres
fig, ax=plt.subplots(2,2)
sns.countplot(x=New_Ethnicity_Comedy, hue=New_Ethnicity_Comedy,data=Comedy_Genre,palette=sns.color_palette("dark", 20),ax=ax[0,0])
ax[0,0].set_title('Top 4 Origin/Ethnicities of Drama Genre')
ax[0,0].set_xlabel('Origin/Ethnicities',fontsize=15)
ax[0,0].set_ylabel('No of movies of Drama Genre',fontsize=15)
sns.countplot(x=New_Ethnicity_Drama, hue=New_Ethnicity_Drama,data=Drama_Genre,palette=sns.color_palette("dark", 20),ax=ax[0,1])
ax[0,1].set_title('Top 4 Origin/Ethnicities of Comedy Genre')
ax[0,1].set_xlabel('Origin/Ethnicities',fontsize=15)
ax[0,1].set_ylabel('No of movies of Comedy Genre',fontsize=15)
sns.countplot(x=New_Ethnicity_Horror, hue=New_Ethnicity_Horror,data=Horror_Genre,palette=sns.color_palette("dark", 20),ax=ax[1,0])
ax[1,0].set_title('Top 4 Origin/Ethnicities of Horror Genre')
ax[1,0].set_xlabel('Origin/Ethnicities',fontsize=15)
ax[1,0].set_ylabel('No of movies of Horror Genre',fontsize=15)
sns.countplot(x=New_Ethnicity_Action, hue=New_Ethnicity_Action,data=Action_Genre,palette=sns.color_palette("dark", 20),ax=ax[1,1])
ax[1,1].set_title('Top 4 Origin/Ethnicities of Action Genre')
ax[1,1].set_xlabel('Origin/Ethnicities',fontsize=15)
ax[1,1].set_ylabel('No of movies of Action Genre',fontsize=15)
```

```python
#creates a 2 by 2 subplot for the directors based on the genres
fig, ax=plt.subplots(2,2)
sns.countplot(New_Directors_Drama, order=pd.value_counts(New_Directors_Drama)[:5].index, palette=sns.color_palette("seismic", 20),ax=ax[0,0])
ax[0,0].set_title('Top 5 Directors for Drama Genre')
ax[0,0].set_xlabel('Directors',fontsize=15)
ax[0,0].set_ylabel('No of movies produced',fontsize=15)
sns.countplot(New_Directors_Comedy, order=pd.value_counts(New_Directors_Comedy)[:5].index, palette=sns.color_palette("seismic", 20),ax=ax[0,1])
ax[0,1].set_title('Top 5 Directors for Comedy Genre')
ax[0,1].set_xlabel('Directors',fontsize=15)
ax[0,1].set_ylabel('No of movies produced',fontsize=15)
sns.countplot(New_Directors_Horror, order=pd.value_counts(New_Directors_Horror)[:5].index, palette=sns.color_palette("seismic", 20),ax=ax[1,0])
ax[1,0].set_title('Top 5 Directors for Horror Genre')
ax[1,0].set_xlabel('Directors',fontsize=15)
ax[1,0].set_ylabel('No of movies produced',fontsize=15)
sns.countplot(New_Directors_Action, order=pd.value_counts(New_Directors_Action)[:5].index, palette=sns.color_palette("seismic", 20),ax=ax[1,1])
ax[1,1].set_title('Top 5 Directors for Action Genre')
ax[1,1].set_xlabel('Directors',fontsize=15)
ax[1,1].set_ylabel('No of movies produced',fontsize=15)
plt.show()
```
